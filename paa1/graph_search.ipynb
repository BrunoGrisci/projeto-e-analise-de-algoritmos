```python
def init_visitados(g):
    """Dicionário mapeando todo nodo para False (não-visitado)."""
    return {n : False for n in g.keys()}
```

## Breadth-First Search (BFS)

```python
def bfs(g, s, visitados):
    """Busca em Largura a partir de um nodo s."""
    queue = deque([s])
    visitados[s] = True

    while len(queue) > 0:
        n = queue.popleft()
        print(n)
        for vizinho in g[n]:
            if not visitados[vizinho]:
                queue.append(vizinho)
                visitados[vizinho] = True
```

```python
bfs(g1, 'A', init_visitados(g1))
```

## Depth-First Search (DFS)

```python
def dfs(g, s, visitados):
    """Busca em profundidade a partir de um nodo s."""
    stack = [s]
    visitados[s] = True
    
    while len(stack) > 0:
        n = stack.pop(-1)
        print(n)
        for vizinho in g[n]:
            if not visitados[vizinho]:
                stack.append(vizinho)
                visitados[vizinho] = True
```

```python
dfs(g1, 'A', init_visitados(g1))
```

```python
def dfs_rec(g, s, visitados):
    """Busca em profundidade a partir de um nodo s (versão recursiva)."""
    visitados[s] = True
    print(s)
    for v in g[s]:
        if not visitados[v]:
            dfs_rec(g, v, visitados)
```

```python
dfs_rec(g1, 'A', init_visitados(g1))
```

## Árvore de Busca

```python
def dfs_iter_tree(g, v):
  """Árvore de Busca de DFS. Retorna dicionarário mapeando cada nodo a seu nodo pai na busca."""
  parent  = {i : i for i in g.keys()}   # arvore de antecessor
  visited = {i : False for i in g.keys()}
  p = [v]               
  visited[v] = True
  
  while len(p) > 0:     
    h = p.pop(0)      
    for i in g[h]:          
      if not visited[i]:   
        p.insert(0,i)       
        parent[i] = h   # marca h como antecessor de i
        visited[i] = True     

  return parent  # devolve a arvore
```

```python
def print_path(tree, source, target):
    """Dado uma árvore de busca, o nodo origem usado para construir a árvore
    e um nodo destino, retorna uma lista com o caminho na árvore."""
    node = target
    path = [target]
    while node != tree[node]:
        node = tree[node]
        path.append(node)
    print(list(reversed(path)))
```

## Cálculo de Distâncias em Grafos Não-Valorados

```python
def bfs_distance(g, s, visitados):
    """Distância (em número de arestas) de s para todos os nodos usando BFS."""
    queue = deque([s])
    visitados[s] = True
    distancias = visitados.copy()
    distancias[s] = 0
    while len(queue) > 0:
        n = queue.popleft()
        for vizinho in g[n]:
            if not visitados[vizinho]:
                queue.append(vizinho)
                visitados[vizinho] = True
                distancias[vizinho] = distancias[n] + 1
    return distancias
```

## Componentes Conexos em Grafo Simples

```python
def components(g):
    """Componentes conexos em grafos simples via DFS."""
    def dfs(g, node, visited, comp, c):
        visited[node] = True
        comp[node] = c
        for v in g[node]:
            if not visited[v]:
                dfs(g, v, visited, comp, c)

    comp = {}
    c = 0
    visited = {i : False for i in g.keys()}
    for n in g.keys():
        if not visited[n]:
            dfs(g, n, visited, comp, c)
            c += 1

    return comp
```

## Teste de Bipartição

```python
def bipartite(g):
    """Teste de bipartição usando BFS."""
    visitados = {i : False for i in g.keys()}
    cor = visitados.copy()  # Mapeia cada nodo para uma cor (0 ou 1)
    for i in g.keys():
        queue = deque([i])
        while len(queue) > 0:
            n = queue.popleft()
            for vizinho in g[n]:
                if not visitados[vizinho]:
                    queue.append(vizinho)
                    visitados[vizinho] = True
                    cor[vizinho] = 1 - cor[n]  # Cada vizinho recebe cor oposta
                else:
                    if cor[n] == cor[vizinho]:  # Se um vizinho já foi visitado e tem mesma cor, forma um ciclo ímpar
                        return False

    return cor
```

## Ordenamento Topológico em Dígrafo Acíclico

```python
def toposort_kahn(g):
  """Ordenamento topológico para DAGs usando BGS (Algoritmo de Kahn)."""
  inDeg = {i: 0 for i in g.keys()}  # graus de entrada

  for i in g.keys():     # para todo nodo i
    for j in g[i]:       # para todo j vizinho de i
      inDeg[j] += 1      # +1 ao grau de entrada de j

  q = deque()            # lista de nodos fonte (source)
  order = []                 # lista de saida (ordenamento)
  for i in g.keys():     
    if inDeg[i] == 0:
      q.append(i)    # inicializa q com nodos de grau 0  

  while len(q) > 0:
    h = q.popleft()     # extrai h de q
    order.append(h)      # insere h em l
    for i in g[h]:   # para todos vizinho i de h
      inDeg[i] -= 1  # decrementa grau de i
      if inDeg[i] == 0:   # se grau chega a zero
        q.append(i)   # insere na lista de nodos fonte
  return order          # retorna a lista de nodos
```

```python
def toposort(g):
    """Ordenamento Topológico via DFS."""
    visitados = init_visitados(g)
    order = []

    def dfs_rec(g, s, visitados, order):
        visitados[s] = True
        for v in g[s]:
            if not visitados[v]:
                dfs_rec(g, v, visitados, order)
        order.append(s)
    
    for i in g.keys():
        if not visitados[i]:
            dfs_rec(g, i, visitados, order)
    
    return order[::-1]  # reverte a lista
```

## Componentes Fortemente Conexos em Dígrafo

```python
def reverse_graph(g):
    """Dado um dígrafo, retorna um dígrafo com a direção de todas as arestas invertidas."""
    rev_g = {i : [] for i in g.keys()}
    for n, vizinhos in g.items():
        for v in vizinhos:
            rev_g[v].append(n)
    return rev_g
```

```python
def kosaraju_shamir(g):
    """Algoritmo de Kosaraju-Shamir para encontrar Componentes Fortemente Conexos em um Dígrafo."""
    rev_g = reverse_graph(g)
    order = toposort(rev_g)

    visited = init_visitados(g)
    components = {}
    c = 0

    def dfs(g, s, visitados, components, c):
        visitados[s] = True
        components[s] = c
        for v in g[s]:
            if not visited[v]:
                dfs(g, v, visited, components, c)

    for i in order:
        if not visited[i]:
            dfs(g, i, visited, components, c)
            c += 1

    return components
```

## Ciclo em Grafo Simples

```python
def has_cycle(g):
    """Detecção de Ciclo em Grafo simples via DFS."""
    visited = {i : False for i in g.keys()}

    def dfs(node, parent):
        visited[node] = True

        for neighbor in g[node]:
            if not visited[neighbor]:
                if dfs(neighbor, node):
                    return True
            elif neighbor != parent:
                return True

        return False

    for i in g.keys():
        if not visited[i]:
            if dfs(i, -1):
                return True

    return False
```

## Ciclo em Dígrafo

```python
def has_cycle_kahn(g):
    """Detecção de Ciclo em Dígrafo via Algoritmo de Kahn."""
    order = toposort_kahn(g)
    if len(order) != len(g.keys()):
        return True
    else:
        return False
```
